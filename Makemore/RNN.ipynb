```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt # for making figures
%matplotlib inline

# Imports
```

```python
# read in all the words
words = open('names.txt', 'r').read().splitlines()
words[:8] # first eight words
```

```python
# build the vocabulary of characters and mappings to/from integers
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(itos)
print(itos)
print(vocab_size)
```

```python
# build the dataset
block_size = 3 # context length: how many characters do we take to predict the next one?

def build_database(words):
    X, Y = [], []

    for w in words:
        context = [0] * block_size
        for ch in w + '.':
            ix = stoi[ch]
            X.append(context)
            Y.append(ix)
            context = context[1:] + [ix] # crop and append

    X = torch.tensor(X)
    Y = torch.tensor(Y)
    print(X.shape, Y.shape)
    return X, Y

import random
random.seed(42)
random.shuffle(words)
n1 = int(0.8*len(words))
n2 = int(0.9*len(words))

# Three splits, Train, Dev, Test
Xtr, Ytr = build_database(words[:n1])       # 80%
Xdev, Ydev = build_database(words[n1:n2])   # 10%
Xte, Yte = build_database(words[n2:])       # 10%
```

```python
# MLP revisited
n_embd = 10 # the dimensionality of the character embedding vectors
n_hidden = 200 # the number of neurons in the hidden layer of the MLP

g = torch.Generator().manual_seed(2147483647) # for reproductibiloty
C = torch.randn((vocab_size, n_embd),               generator=g)
W1 = torch.randn((n_embd * block_size, n_hidden),   generator=g)* (5/3)/((n_embd * block_size)**0.5) #0.2
# b1 = torch.randn(n_hidden,                          generator=g) * 0.01
# it's okay to set the biases to very small number to get a little bit of entrophy
W2 = torch.randn((n_hidden, vocab_size),            generator=g) * 0.01 # scale down W2
# Don't set W's or weights to be zeroes,
b2 = torch.randn(vocab_size,                        generator=g) * 0 # b2 at 0 at initialization

# batch normalization gain 
bngain = torch.ones((1, n_hidden))
bnbias = torch.zeros((1, n_hidden))
bnmean_running = torch.zeros((1, n_hidden))
bnstd_running = torch.ones((1, n_hidden))

parameters = [C, W1, W2, b2, bngain, bnbias]
print(sum(p.nelement() for p in parameters)) # number of parameter in total
for p in parameters:
    p.requires_grad = True
```

```python
x = torch.randn(1000, 10) # example, fan_in = 10
# 1000 examples, 10 dimensions
# some random input x that is drawn from a gaussian
w = torch.randn(10, 200) / 10**0.5
# divide by input square root
# multiplied by a bigger number, standard deviation will scale
# vice versa by a smaller number
# what do you multiply to get 1?

# 10 inputs, 200 neurons in the hidden layer
# initialized using gaussian

y = x @ w # multiplication to get the preactivations of the neurons
# the analysis here looks at,
# basically both units are from gaussian, if x @ w, and forget bias and non-linearity
# then what is the mean and standard deviation of these gaussian?

# In the beginning, it's just a normal gaussian distribution,
# mean=0, standard deviation=1
# standard deviation = measure the spread of the gaussian 

print(x.mean(), x.std())
print(y.mean(), y.std())
# expanding the gaussian from the input
# we don't want that, 
# we want most of the neural nets to have relatively similar actiations
# unit gaussian roughly throw off the neural nets

plt.figure(figsize=(20,5))
plt.subplot(121)
plt.hist(x.view(-1).tolist(), 50, density=True);
plt.subplot(122)
plt.hist(y.view(-1).tolist(), 50, density=True);
```

```python
# same optimiaztion as last time
max_steps = 200000
batch_size = 32
lossi = []

# Cleaned up on magic numbers
for i in range(max_steps):

    # minibatch construct
    ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
    Xb, Yb = Xtr[ix], Ytr[ix] # batch X,Y

    # forward pass
    emb = C[Xb] # embed the characters into vectors
    embcat = emb.view(emb.shape[0], -1) # concatenate the vectors
    
    # Linear layer
    hpreact = embcat @ W1 # + b1 # hidden layer pre-activation
    # we want hpreact closer to zero
    # normalize preact
    # hpreact should be roughly gassian
    # hpreact, subtract the mean then divide by std
    # every single neuron and its firing rate will be eactly unit gaussian on the 32 examples atleast
    # batch normalization, all perfectably differentiable
    # we only want these to be gaussian to be only at initialization
    # multiply by bngain and offset by bnbias

    # BatchNorm layer
    # -------------------------------------------------------
    bnmeani = hpreact.mean(0, keepdim=True)
    bnstdi = hpreact.std(0, keepdim=True)

    hpreact = bngain * (hpreact - bnmeani) / bnstdi + bnbias
    # batch normalization comes with a terrible cost
    # it used to be, a single example feeding into the neural net,
    # calculate its activation and logits, so you arrive at some logits for this exmaples
    # bc of the normalization through the batch, we are coupling these examples mathmetically

    with torch.no_grad(): # not building out a graph
        # being updated not using gradient descent
        bnmean_running = 0.999 * bnmean_running + 0.001 * bnmeani
        bnstd_running = 0.999 * bnstd_running + 0.001 * bnstdi
    # -------------------------------------------------------
    # Non-linearity
    h = torch.tanh(hpreact) # hidden layer
    # torch.tanh function, it takes arbituary numbers and squashes them into -1, 1 smoothly
    logits = h @ W2 + b2 # output layer, we want this closer to zero
    loss = F.cross_entropy(logits, Yb) # loss function

    # backward pass
    for p in parameters:
        p.grad = None
    loss.backward()

    # update
    lr = 0.1 if i < 100000 else 0.01 # step learning rate decay
    for p in parameters:
        p.data += -lr * p.grad

    # track stats
    if i % 10000 == 0: # print every once in a while
        print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
    lossi.append(loss.log10().item())


# Loss initialization is what we expect
# this is only one layer MLP
# so the fix is simple

# Batch normalization happens to have the regularization efefct, and stablize training
# It works well bc of the regularization effect, and bc it is quite effectective at controlling the activation and distribution
```

```python
@torch.no_grad() # this decorator disables gradient tracking
# Whatever happens in this whole function is assumed by torch to never require any gradients
# does not anticipate .backward()

def split_loss(split): # Pass in a string on which split you like
    x,y = {
        'train': (Xtr, Ytr),
        'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x] # (N, block_size, n_embd)
    embcat = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    
    # normalize then scale
    hpreact = embcat @ W1 # + b1
    # hpreact = bngain * (hpreact - hpreact.mean(0, keepdim=True)) / hpreact.std(0, keepdim=True) + bnbias
    hpreact = bngain * (hpreact - bnmean_running) / bnstd_running + bnbias
    h = torch.tanh(hpreact) # (N, n_hidden)
    logits = h @ W2 + b2 # (N, vocab_size)
    loss = F.cross_entropy(logits, y)
    print(split, loss.item())

split_loss('train')
split_loss('val')

# The loss is closer to train a val since we initialize our NN better
# Instead of spending the first several thousands tries to squash down the the loss
# It uses all the iterations to train the NN instead
```

```python
# Let's train a deeper network
# The classes we create here are the same API as nn.Module in PyTorch

class Linear:

    # fan_in num of inputs
    # fan_out num of outputs
    # bias=True whether or not we want to use a bias
    def __init__(self, fan_in, fan_out, bias=True):
        # kami initialization
        self.weight = torch.randn((fan_in, fan_out), generator=g) / fan_in**0.5
        self.bias = torch.zeros(fan_out) if bias else None

    def __call__(self, x):
        # w * x + b
        self.out = x @ self.weight
        if self.bias is not None:
            self.out += self.bias
        return self.out

    def parameters(self):
        return [self.weight] + ([] if self.bias is None else [self.bias])
# LINEAR LAYER

class BatchNorm1d:

    def __init__(self, dim, eps=1e-5, momentum=0.1):
        self.eps = eps
        self.momentum = momentum
        self.training = True
        # parameters (trained with backprop)
        self.gamma = torch.ones(dim)
        self.beta = torch.zeros(dim)
        # buffers (trained with a running 'momentum update')
        self.running_mean = torch.zeros(dim)
        self.running_var = torch.ones(dim)

    def __call__(self, x):
        # calculate the forward pass
        if self.training:
            xmean = x.mean(0, keepdim=True) # batch mean
            xvar = x.var(0, keepdim=True) # batch variance
        else:
            xmean = self.running_mean
            xvar = self.running_var
        xhat = (x - xmean) / torch.sqrt(xvar + self.eps) # normalize to unit variance
        self.out = self.gamma * xhat + self.beta
        # .out is used for making graphs, normally theres no .out
        # update the buffers
        if self.training:
            with torch.no_grad():
                self.running_mean = (1 - self.momentum) * self.running_mean + self.momentum * xmean
                self.running_var = (1 - self.momentum) * self.running_var + self.momentum * xvar
        return self.out
    
    def parameters(self):
        return [self.gamma, self.beta]
# BATCH NORMALIZATION LAYER

class Tanh:
    def __call__(self, x):
        self.out = torch.tanh(x)
        return self.out
    def parameters(self):
        return []

n_embd = 10 # the dimensionality of the character embedding vectors
n_hidden = 100 # the number of neurons in the hidden layer of the MLP
g = torch.Generator().manual_seed(2147483647) # for reproducibility

C = torch.randn((vocab_size, n_embd),            generator=g)
layers = [
    Linear(n_embd * block_size, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(           n_hidden, n_hidden, bias=False), BatchNorm1d(n_hidden), Tanh(),
    Linear(           n_hidden, vocab_size, bias=False), BatchNorm1d(vocab_size),
]
# layers = [
#   Linear(n_embd * block_size, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, n_hidden), Tanh(),
#   Linear(           n_hidden, vocab_size),
# ]

with torch.no_grad():
    # last layer: make less confident
    layers[-1].gamma *= 0.1
    #layers[-1].weight *= 0.1
    # all other layers: apply gain
    for layer in layers[:-1]:
        if isinstance(layer, Linear):
            layer.weight *= 1.0 #5/3

parameters = [C] + [p for layer in layers for p in layer.parameters()]
print(sum(p.nelement() for p in parameters)) # number of parameters in total
for p in parameters:
    p.requires_grad = True
```

```python
# same optimization as last time
max_steps = 200000
batch_size = 32
lossi = []
ud = []

for i in range(max_steps):

  # minibatch construct
  ix = torch.randint(0, Xtr.shape[0], (batch_size,), generator=g)
  Xb, Yb = Xtr[ix], Ytr[ix] # batch X,Y
  
  # forward pass
  emb = C[Xb] # embed the characters into vectors
  x = emb.view(emb.shape[0], -1) # concatenate the vectors
  for layer in layers:
    x = layer(x)
  loss = F.cross_entropy(x, Yb) # loss function
  
  # backward pass
  for layer in layers:
    layer.out.retain_grad() # AFTER_DEBUG: would take out retain_graph
  for p in parameters:
    p.grad = None
  loss.backward()
  
  # update
  lr = 0.1 if i < 150000 else 0.01 # step learning rate decay
  for p in parameters:
    p.data += -lr * p.grad

  # track stats
  if i % 10000 == 0: # print every once in a while
    print(f'{i:7d}/{max_steps:7d}: {loss.item():.4f}')
  lossi.append(loss.log10().item())
  with torch.no_grad():
    ud.append([((lr*p.grad).std() / p.data.std()).log10().item() for p in parameters])

  if i >= 1000:
    break # AFTER_DEBUG: would take out obviously to run full optimization
```

```python
@torch.no_grad() # this decorator disables gradient tracking
def split_loss(split):
    x,y = {
        'train': (Xtr, Ytr),
            'val': (Xdev, Ydev),
        'test': (Xte, Yte),
    }[split]
    emb = C[x] # (N, block_size, n_embd)
    x = emb.view(emb.shape[0], -1) # concat into (N, block_size * n_embd)
    for layer in layers:
        x = layer(x)
    loss = F.cross_entropy(x, y)
    print(split, loss.item())

# put layers into eval mode
for layer in layers:
    layer.training = False
split_loss('train')
split_loss('val')
```
